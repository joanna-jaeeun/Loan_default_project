```python
# Check total null values in the dataframe
df.isna().sum()[1:20], df.isna().sum()[20:40], df.isna().sum()[41:60], df.isna().sum()[61:80]

def print_df(df, chunk_size = 20) :
    na_series = df.isna().sum()
    total  = len(na_series)

    for start in range(1, total, chunk_size) :
        end = start + chunk_size
        print(na_series[start : end])
        print("_" * 40)

print_df(df, chunk_size = 20)
```
